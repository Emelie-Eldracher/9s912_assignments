```python
from pytube import YouTube
from youtube_transcript_api import YouTubeTranscriptApi
import json
import time

# get list of videos so far
import os

directory_path_videos = '/content/drive/MyDrive/The Football Archive™/Videos/'
directory_videos = os.listdir(directory_path_videos)

created_video_ids = set(i.split(' - ')[-1][:-4] for i in directory_videos)
created_video_ids

directory_path_transcripts = '/content/drive/MyDrive/The Football Archive™/Transcripts/'
directory_transcripts = os.listdir(directory_path_transcripts)

created_transcript_ids = set(i.split(' - ')[-1][:-5] for i in directory_transcripts)
created_transcript_ids

def Download(link, video=True, transcript=True):

    video_id = link.split("?v=")[1]

    # check if video already downloaded
    if video and transcript and video_id in created_video_ids and video_id in created_transcript_ids:
        print(f"{video_id} already downloaded", end="\r")
        return

    if video and video_id in created_video_ids:
        print(f"{video_id} already downloaded", end="\r")
        return

    if transcript and video_id in created_transcript_ids:
        print(f"{video_id} already downloaded", end="\r")
        return

    youtubeObject = YouTube(link)
    title = youtubeObject.title
    author = youtubeObject.author
    publish_date = youtubeObject.publish_date.strftime("%Y-%m-%d")

    title = f"{author} - {publish_date} - {title} - {video_id}".replace("/", "-")
    print(f"{title}")

    if video:
        print(f"Downloading video...")
        start_time = time.time()
        # dowload the video
        youtubeObject = youtubeObject.streams.get_by_resolution('720p')
        print(f"Approx size: {youtubeObject.filesize_gb}GB")
        youtubeObject.download(output_path="/content/drive/MyDrive/The Football Archive™/Videos/", filename=f"{title}.mp4")
        print(f"Finished Downloading in {time.time() - start_time} seconds")

    if transcript:
        if video_id in created_transcript_ids:
            print(f"Transcript {video_id} already downloaded")
            return
        print(f"Downloading transcript...")
        # assigning srt variable with the list of dictionaries
        # obtained by the .get_transcript() function
        # and this time it gets only the subtitles that
        # are of english language.
        srt = YouTubeTranscriptApi.get_transcript(video_id,
                                                languages=['en'])

        # save as json
        with open(f'/content/drive/MyDrive/The Football Archive™/Transcripts/{title}.json', 'w') as fout:
            json.dump(srt, fout)

# link = "https://www.youtube.com/watch?v=gVdVao2e7SI"
# Download(link, video=False, transcript=False)

channel_dict = {
    "QBSchool": "https://www.youtube.com/channel/UCGfxHp2yuimtLsFaq1r7aIQ",
    "Kurt Warner": "https://www.youtube.com/c/kurtwarnerqbc",
    # "Xs and Noles": "https://www.youtube.com/channel/UC6yCZXabzBP_HUfEl8L7BaA",
    # "The DB Room": "https://www.youtube.com/c/thedbroom",
    "Coach Dan Casey": "https://www.youtube.com/c/CoachDanCasey",
    "Football Analysis": "https://www.youtube.com/channel/UCjoHPnySKcDbxzaF2R8YM2Q",
}

from pytube import Channel

for channel_name, channel_link in channel_dict.items():

    # get all video links for a channel
    c = Channel(channel_link)
    video_urls = c.video_urls
    # print(video_urls[:3])

    for url in video_urls:
        try:
            Download(url, video=False, transcript=True)
        except Exception as e:
            print(e.CAUSE_MESSAGE)
```

```python
# get transcripts from the json files

# Step 1: Open json
directory_path_transcripts = '/content/drive/MyDrive/The Football Archive™/Transcripts/'
directory_transcripts = os.listdir(directory_path_transcripts)

directory_path_txts = '/content/drive/MyDrive/The Football Archive™/Transcripts/Texts/'
directory_txts = os.listdir(directory_path_txts)

total_words = 4211417
new_words = 0

for file in directory_transcripts:
    if f'{file[:-5]}.txt' in directory_txts:
        continue

    if file == "Texts":
        continue

    try:
        with open(f'/content/drive/MyDrive/The Football Archive™/Transcripts/{file}', 'r') as f:
            data = json.load(f)
    except:
        print(file)
        continue

    transcript_txt = ' '.join(line['text']for line in data)
    script_words = len(transcript_txt.split())
    new_words += script_words
    print(f'{script_words}: {file[:-5]}')

    # save txt as .txt file
    with open(f'/content/drive/MyDrive/The Football Archive™/Transcripts/Texts/{file[:-5]}.txt', 'w') as f:
        f.write(transcript_txt)

print(f"NEW WORDS: {new_words}")
print(f"TOTAL WORDS: {total_words+new_words}")

```
