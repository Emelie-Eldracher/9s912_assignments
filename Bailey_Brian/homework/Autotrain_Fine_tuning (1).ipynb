##Using the Model

```python
from transformers import AutoTokenizer, LlamaForCausalLM

model = LlamaForCausalLM.from_pretrained("brianbailey18/football_llama_13b_init_trained_chat_model_10_epochs")
tokenizer = AutoTokenizer.from_pretrained("brianbailey18/football_llama_13b_init_trained_chat_model_10_epochs")
```

```python
stop_word = "###"
stop_word_id = tokenizer(stop_word, add_special_tokens=False)['input_ids']
```

```python
from transformers import StoppingCriteriaList, StoppingCriteria
import torch
```

```python
class StoppingCriteriaSub(StoppingCriteria):
    def __init__(self, stops=[], encounters=1):
        super().__init__()
        self.stops = stops
        self.ENCOUNTERS = encounters

    def __call__(self, input_ids: torch.LongTensor, scores: torch.FloatTensor):
        stop_count = 0
        for stop in self.stops:
            stop_count += (input_ids[0] == stop).sum().item()

        if stop_count >= self.ENCOUNTERS:
            return True
        return False
```

```python
stopping_criteria = StoppingCriteriaList([StoppingCriteriaSub(stops=stop_word_id, encounters=2)])
```

```python
prompts = [
    "How many downs do you get to move the ball 10 yards?",
    "How many defensive backs are on the field in a nickel personnel package?",
    "Can you name how many players of each offensive skill position are on the field in 12 personnel?",
    "Can you name how many players of each offensive skill position are on the field in 11 personnel?",
    "What is an RPO?",
    "Can you explain a Cover 3 defense?" ,
    "Can you explain the passing concept 'Mesh'?",
    "Can you explain what a “cloud” coverage technique is?" ,
    "When do teams usually play man coverage?",
    "What QB footwork goes with a Four Verticals passing concept?",
    "What is a bubble screen?",
    "Can you explain how to run Power?",
    "Can you explain the ‘Stick’ concept?",
    "Can you explain the QB’s read in Mesh Rail?",
    "What passing concept usually fits well with inside zone for an RPO?",
    "What are good ways to protect against the blitz?",
    "Can you explain Cover 2 Man?",
    "Can you explain what a ‘Hot 3’ defense means?",
    "How would you stop a team who has a good running back and a bad quarterback?",
    "Can you explain the  ‘Flood’ concept?",
    "Can you explain ‘Jet’ protection?",
    "Can you explain the ‘Levels’ concept?",
    "Can you explain what it means to have a ‘high-low’ read?",
    "Can you explain what an ‘orbit’ motion is?",
    "What is a ‘Jet’ motion?",
]
```

```python
prompts = [
    "What is a ‘Jet’ motion?"
] #testing
```

```python
responses, count = {}, 1

for prompt in prompts:
    inputs = tokenizer(prompt, return_tensors="pt")

    # Generate
    generate_ids = model.generate(inputs.input_ids, max_length=125, stopping_criteria=stopping_criteria)

    response = tokenizer.batch_decode(generate_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False)[0]
    split_response = response.split('\n')[1:]
    responses[prompt] = " ".join(split_response)
    print('count: ', count)
    count += 1
```

```python
file_path = "/content/drive/My Drive/The Football Archive™/Training Data/fine-tuned-responses.txt"

# Open the file in write mode and save the dictionary
with open(file_path, "w") as f:
  for key, value in responses.items():
    f.write(f"{key}: {value}\n")
```

```python
for response in responses:
  print('Prompt : ' , response, '\n')
  print('Response : ', responses[response], '\n************************************')
```
