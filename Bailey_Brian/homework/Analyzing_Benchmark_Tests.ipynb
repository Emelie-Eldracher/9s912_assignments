```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Scale:
# 0-Little to no reference to the sport of football
# 1-Using football terminology but all wrong
# 2-Using football terminology with some correct words
# 3-Using football terminology and mostly correct
# 4-Using football terminology and correct
```

```python
fine_tuned_grades = [4,2,2, 4, 4, 3,4, 1, 4, 3, 3, 2, 1, 1, 3, 4, 3, 2, 4, 2, 2, 4, 3, 1, 4]
plain_grades = [4, 4, 2, 3, 0, 3, 0, 0, 2, 2, 3, 0, 0, 1, 1, 0, 2, 0, 4, 0, 0, 0, 0, 0, 0]
```

```python
# Calculate average grades and SEM (Standard Error of the Mean)
fine_tuned_avg = np.mean(fine_tuned_grades)
plain_avg = np.mean(plain_grades)
fine_tuned_sem = np.std(fine_tuned_grades) / np.sqrt(len(fine_tuned_grades))
plain_sem = np.std(plain_grades) / np.sqrt(len(plain_grades))

# Create a DataFrame for Seaborn
import pandas as pd
df = pd.DataFrame({
    'Model': ['Fine-tuned', 'Plain'],
    'Average Score': [fine_tuned_avg, plain_avg],
    'SEM': [fine_tuned_sem, plain_sem]
})

# Create the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Average Score', data=df, yerr=df['SEM'].values, capsize=0.1)
plt.title('Performance Comparison')
plt.show()
```

```python
# Identify examples for the three cases
both_score_4 = [i for i, (ft, pl) in enumerate(zip(fine_tuned_grades, plain_grades)) if ft == 4 and pl == 4]
fine_tuned_4_plain_0 = [i for i, (ft, pl) in enumerate(zip(fine_tuned_grades, plain_grades)) if ft == 4 and pl == 0]
neither_high = np.argmin(np.array(fine_tuned_grades) + np.array(plain_grades))

print("Indices where both models scored 4:", both_score_4)
print("Indices where fine-tuned model scored 4 and plain model scored 0:", fine_tuned_4_plain_0)
print("Index where neither model scored highly:", neither_high)
```
